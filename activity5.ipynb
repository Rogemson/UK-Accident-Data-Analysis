<h3> Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3> Importing dataframe</h3>

```python
df = pd.read_csv('internet_demography.csv')
```

```python
df
```

<h2> 1. Histogram: Distribution of Birth Rates

```python
plt.figure(figsize=(15, 6))
sns.histplot(df["Birth rate"], bins=5, kde=True)
plt.xlabel("Birth Rate")
plt.ylabel("Frequency")
plt.title("Distribution of Birth Rates")
plt.show()
```

<h2> 2. Top 5 Countries with Highest Internet Usage

```python
top_5 = df.nlargest(5, "Internet users")
plt.figure(figsize=(15, 6))
sns.barplot(x = "Country Name", y = "Internet users", data = top_5)
plt.xlabel("Country")
plt.ylabel("Internet Usage (%)")
plt.title("Top 5 Countries with Highest Internet Usage")
plt.show()
```

<h2>3. Count of Countries by Income Group

```python
order = ["Low income", "Middle income", "High income"]

plt.figure(figsize=(15, 6))
sns.countplot(x=df["Income Group"], order=order)
plt.xlabel("Income Group")
plt.ylabel("Count")
plt.title("Number of Countries by Income Group")
plt.show()
```

<h2> 4. Box Plot of Internet Usage

```python
plt.figure(figsize=(15, 6))
sns.boxplot(y=df["Internet users"])
plt.ylabel("Internet users")
plt.title("Box Plot of Internet Usage")
plt.tight_layout()
plt.show()
```

<h2> 5. Countries with No Internet Access (None in this dataset)

```python
no_internet = df[df["Internet users"] == 0]
print("Countries with No Internet Access:")
print(no_internet)
```

<h2> 6. Average Birth Rate

```python
avg_birth_rate = df["Birth rate"].mean()
print(f"Average Birth Rate: {avg_birth_rate:.2f}")
```

<h2> 7. Scatter Plot: Birth Rate vs. Internet Usage

```python
plt.figure(figsize=(15, 6))
sns.scatterplot(x="Birth rate", y="Internet users", data=df)
plt.xlabel("Birth Rate")
plt.ylabel("Internet Usage (%)")
plt.title("Scatter Plot: Birth Rate vs. Internet Usage")
plt.show()
```

<h2> 8. Countries with Birth Rate Above 40

```python
high_birth_rate = df[df["Birth rate"] > 40]
print("Countries with Birth Rate Above 40:\n", high_birth_rate)
```

<h2>9. Pie Chart of Income Groups

```python
plt.figure(figsize=(15, 6))
df["Income Group"].value_counts().plot.pie(autopct='%1.1f%%', colors=["red", "blue", "green"])
plt.title("Proportion of Countries by Income Group")
plt.ylabel("")
plt.show()
```

<h2> 10. Correlation Between Birth Rate and Internet Usage

```python
correlation = df["Birth rate"].corr(df["Internet users"])
print(f"Correlation between Birth Rate and Internet Usage: {correlation:.2f}")
```
