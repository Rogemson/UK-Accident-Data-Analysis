Discretionary spending here is defined as the sum of non-essential categories‚Äîdining out, alcohol, tobacco, special occasions, and miscellaneous goods/services. This shows which age groups splurge most on ‚Äúwants‚Äù rather than necessities.

| Age Group | Avg. Discretionary Spend (‚Ç±) |

| 40‚Äì49     | 37,949                        |

#### Key Insights

- **Younger and Older Frugality**

```python
discretionary_cats = [
    'Restaurant and hotels Expenditure',
    'Alcoholic Beverages Expenditure',
    'Tobacco Expenditure',
    'Miscellaneous Goods and Services Expenditure',
    'Special Occasions Expenditure'
]

age_bins = range(20, 81, 10)
age_labels = [f"{b}‚Äì{b+9}" for b in age_bins[:-1]]
df['Age Group'] = pd.cut(df['Household Head Age'],bins=age_bins,right=False,labels=age_labels)

df['Discretionary Spend'] = df[discretionary_cats].sum(axis=1)

age_disc = (df.groupby('Age Group')['Discretionary Spend'].mean().reset_index().sort_values('Discretionary Spend', ascending=False))

plt.figure(figsize=(15, 6))
sns.barplot(data=age_disc, x='Age Group', y='Discretionary Spend', palette='magma')
plt.title('Average Discretionary Spending by Head‚Äôs Age Group')
plt.xlabel('Age Group')
plt.ylabel('Avg. Discretionary Spend (‚Ç±)')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
